Dynamic Time Warping

With dynamic time warping, we can compare two signals which are not aligned in the time domain and determine the shortest connecting path. This technique is commonly used in applications such as speech recognition where target frequencies naturally occur at different time points and for various durations in audio. It allows us to stretch or compress the time axis of one sequence to align with the other sequence with the determined distance. This warping process is dynamic because it considers all possible alignments and selects the one with the minimum total cost. Here we will explore the code to calculate the shortest path between two matching signals offset in time and visualize the alignment.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib as mpl
import matplotlib.pyplot as plt

from scipy import signal
from scipy.spatial import distance_matrix
from scipy.spatial import distance
```

We start by creating two example sinusoidal signals offset in the time domain. We will attempt to match the corresponding peaks and troughs of the signals through dynamic time warping. 

```python
## Sine waves
sr = 3
ts = 1.0/sr
t = np.arange(0,10,ts)
n = np.random.normal(scale=1, size=t.size)
y = 2*np.sin(2*np.pi*0.2*t) 
y += 1*np.cos(2*np.pi*0.5*t)
y += 3*np.sin(2*np.pi*0.7*t) + n

n = np.random.normal(scale=0.01, size=t.size)
space = 0.5*np.sin(2*np.pi*0.1*t) 
space += 0.2*np.cos(2*np.pi*0.1*t)
space += 0.3*np.sin(2*np.pi*0.2*t+4) + n

#Create two sine waves
y1 = np.append(space[:len(space)//2],y)
y1 = np.append(y1, space)
y2 = np.append(space, y)
y2 = np.append(y2, space[:len(space)//2])

#Plot
plt.rcParams['figure.figsize'] = [14, 10] 
plt.subplot(211)
plt.plot(np.arange(0,75,1/y1[0].size), y1, 'purple')
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
plt.subplot(212)
plt.plot(np.arange(0,75,1/y2[0].size), y2, 'green')
plt.xlabel('t (s)')
plt.ylabel('Amplitude')
```

Dynamic time warping algorithm:


- Arrange two sequences, X and Y, in an n x m matrix with position (i, j) corresponding to the ith and jth values 
  of X and Y respectively.

- Create a distance matrix, where each element represents the cost of aligning a point from the first sequence
  with a point from the second sequence. The cost can be a distance measure, such as Euclidean distance or 
  similarity measures.

- Initialize a dynamic programming cost matrix with the same dimensions as the distance matrix. Each element in the 
  dynamic programming matrix represents the accumulated cost of the best alignment up to that point.

- Compute the accumulated cost matrix by iteratively filling in the elements. Start from the top-left element 
  and work row by row or column by column, calculating the cumulative cost by considering the costs of neighboring 
  elements. The cost for each element is the sum of the cost of the current alignment and the minimum cost from the 
  three adjacent elements (top, left, and diagonal). The minimum cost ensures that the alignment follows the optimal path.

- Once the cost matrix is fully computed, the minimum cost path is traced back from the bottom-right 
  element to the top-left element. This path represents the best alignment between the two sequences.

- The similarity or distance measure between the two sequences is the value in the bottom-right element of the 
  dynamic programming matrix or the accumulated cost along the minimum cost path.
  
  


We will implement dynamic time warping with a class that determines the distance, cost, and path matrices and stores them as attributes. This will allow us to easily retrieve them as needed. To calculate the initial distance matrix of the two signals we created, we will make use of SciPy's distance matrix functions. To calculate the distance between 2-dimensional arrays, scipy's 'spatial.distance.cdist' method is used. Otherwise, 'spatial.distance_matrix' is used. We can specify which method to use by passing the 'cdist' parameter to our class. 

Continuing, we define our dynamic programming cost matrix and path methods. Following the dynamic time warping algorithm, we create an appropriately shaped cost matrix, fill in the values of the first row and the first column with infinity, set (1,1) to 0, and begin filling in the cost matrix with distance matrix values and minimal costs of neighbors when we determine the path move. We keep track of our path by appending each optimal move choice to a path array. 

To visualize the alignment, we create an additional class method which plots the values of the original arrays as well as the matching path we determined. We will plot the array values normally and draw lines between them revealing the dynamic time warping alignment. If desired, we can stylize our plot by applying a colormap to the path based on segment length. We do this by creating a path object for each pair of matched indices which tracks distance values (we use the pythagorean theorem!). We then create a generator object for our colormap (here we use 'plasma') through which we can iterate and retrieve colors in sequence. When plotting the lines, we iterate over a list of sorted path objects and retrieve new colors from the generator. This will plot the alignment with a satisfying gradient that is consistent with distances in the alignment. If desired, we can also plot a line representing an approximate average of the signals for visualization purposes.

```python
## Dynamic time warping 

#Class to store ordered costs associated with positions i, j  
class path_cost: #Used to plot color gradient 
    def __init__(self, i, j, x1, x2): 
        self.dist = abs(max(x1) - max(x2))
        self.pos = (i, j)
        self.cost = abs(x1[i] - x2[j]) + 2*self.dist
        self.hyp = np.sqrt(abs(j-i)**2 + self.cost**2)
    def __repr__(self):
        return repr((self.pos, self.hyp))
    
#Dynamic time warping class
class dtw:
    def __init__(self, x1, x2, cdist=False):
        self.x1 = x1
        self.x2 = x2
        self.dist_matrix = np.nan
        self.cdist_matrix = np.nan
        self.cost_matrix = np.nan
        self.path = np.nan
        self.avg = np.nan
        self.alignmnent_cost = np.nan
        if cdist == False:
            self.get_dist_matrix(self.x1, self.x2)
            self.get_cost_matrix(self.dist_matrix)
            self.path_mu(self.x1, self.x2)
            self.plot_path(self.x1, self.x2)
            self.plot_cost_matrix()
        else:
            
            self.get_cdist_matrix(np.array(self.x1), np.array(self.x2))
            self.get_cost_matrix(self.cdist_matrix)
            self.path_mu(self.x1, self.x2)
            self.plot_path(self.x1, self.x2)
            self.plot_cost_matrix()
            
    #Cdist matrix    
    def get_cdist_matrix(self, x1, x2): #For 2D arrays
        cdist_matrix = distance.cdist(abs(x1), abs(x2)) #Cdist matrix
        self.cdist_matrix = cdist_matrix
        
    #Dist matrix    
    def get_dist_matrix(self, x1, x2): #For normal arrays
            dist_matrix = distance_matrix(x1.reshape(-1,1), x2.reshape(-1,1))  
            self.dist_matrix = dist_matrix
            return dist_matrix

    #Cost matrix
    def get_cost_matrix(self, dist_matrix): #Adapted from kamperh github repository
        print(f'dist_matrix:\n\n{dist_matrix}\n\nShape: {dist_matrix.shape}\n')
        #Cost matrix
        N, M = dist_matrix.shape
        cost_matrix = np.zeros((N+1, M+1))
        path_matrix = np.zeros((N, M))
        for i in range(1, N+1):
            cost_matrix[i,0] = np.inf
        for j in range(1, M+1):
            cost_matrix[0,j] = np.inf
        for i in range(N):
            for j in range(M):
                alignment_cost = [
                    cost_matrix[i,j], #Match
                    cost_matrix[i,j+1], #Insertion
                    cost_matrix[i+1,j] #Deletion
                    ]
                cost_matrix[i+1,j+1] = dist_matrix[i,j] + min(alignment_cost)
                path_matrix[i,j] = np.argmin(alignment_cost)
        #Path matrix
        i, j = N-1, M-1
        alignment_path = [(i, j)]
        while i > 0 or j > 0:
            move = path_matrix[i,j]
            if move == 0: #Match
                i = i - 1
                j = j - 1
            elif move == 1: #Insertion
                i = i - 1
            elif move == 2: #Deletion:
                j = j - 1
            alignment_path += [(i, j)]
        #Return path and cost matrix
        cost_matrix = cost_matrix[1:, 1:]
        print(f'Path:\n\n{alignment_path}\n\nCost matrix:\n\n{cost_matrix}\n\nShape: {cost_matrix.shape}\n')
        self.path, self.cost_matrix = alignment_path[::-1], cost_matrix
        self.alignment_cost = self.cost_matrix[-1,-1]
        return (alignment_path[::-1], cost_matrix)
    
    #Calculate geometric mean 
    def get_geo_mean(self, X): #Used in calculation of optional mean values
        n = len(X)
        gm = (X + 1).prod()**(1/n) -1
        return gm 

    #Get average path
    def path_mu(self, x1, x2): #Optional mean path
        x1 = x1.ravel()
        x2 = x2.ravel()
        path = self.path
        mu = [self.get_geo_mean(np.array([x1[i], x2[j]])) for i, j in path]
        self.avg = mu
                
    #Plot
    def plot_path(self, x1, x2):
        x1 = x1.ravel()
        x2 = x2.ravel()
        #Colormap path distance
        dist = abs(max(x1) - min(x2)) 
        cmap = mpl.colormaps['plasma'].resampled(len(self.path)).colors
        cgen = (tuple(i) for i in cmap) #Generator object to return colors in sequence
        cost_objects = [path_cost(i, j, x1, x2) for i, j in self.path] #Path objects to track cost order
        cost_objects = sorted(cost_objects, key=lambda x: x.hyp, reverse=True) #Sort path object order
        plt.figure(figsize=(18,6))
        scale_factor = 1.1
        for obj in cost_objects:
            i = obj.pos[0]
            j = obj.pos[1]
            plt.plot([i*scale_factor, j], [x1[i] + dist, x2[j] - dist], color=next(cgen), alpha=0.7) #Apply color to cost line with generator
        x1_xrange = np.linspace(start=0, stop=scale_factor*(len(x1)-1), num=len(x1))
        x2_xrange = np.arange(start=0, stop=x2.shape[0])
        #mu_stop = scale_factor*(len(x1)-1)
        #mu_range = np.linspace(start=0.5, stop=mu_stop + (x2_xrange[-1]-mu_stop)/2, num=len(self.path))
        plt.plot(x1_xrange, x1 + dist, color=(0.723444, 0.196158, 0.538981, 1.))
        plt.plot(x2_xrange, x2 - dist, color=(0.85975 , 0.360588, 0.406917, 1.))
        #plt.plot(mu_range, self.avg, color=(0.752312, 0.227133, 0.513149, 1.), alpha=0.5, linestyle='--')
        plt.axis('off')
        
    #Plot cost matrix
    def plot_cost_matrix(self):
        #Plot cost matrix
        fig, ax = plt.subplots(figsize=(15, 12))
        ax = sns.heatmap(self.cost_matrix.T, linewidths=0.01, cmap="plasma", ax=ax)
        ax.invert_yaxis()
        #Path
        path_i = [tup[0] + 0.5 for tup in self.path]
        path_j = [tup[1] + 0.5 for tup in self.path]
        ax.plot(path_i, path_j, color=(0.95547 , 0.533093, 0.28549 , 1.), linewidth=3, alpha=1)
```

```python
warped = dtw(y1, y2)
```

We see how the peaks and troughs of each sinusoid are aligned despite being offset in time. The heatmap of the cost matrix representing possible alignment paths shows that the dynamic time warping algorithm we employed did in fact determine the optimal path between the two signals.
